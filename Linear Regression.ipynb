# Supervised & Unsupervised

SUPERVISED LEARNING : is where u have both the input variable X & output variable Y & u see an algo... to learn the mapping func.. fro X to Y & predict the output of new data. Supervise learning can further b classified as a classification or a regression techniques.

What do u mean by regression? when 1 or more than 2 variables fluctuating the value of output is known as regression. In regression one variable is Dependent Variable(Output Variable) & another variable is Independent Variable(Input Variable).

LINEAR REGRESSION:

is a type of supervised machine learning alogo.. thatcomputes the linear relationship b/t a dependent variable/label & 1 or more  independent variable/features(r the non dependent columns).
When the number of the dependent features is one then it is known as univariant Linear regression & in the case of more then one feature it is known as multivariant linear regression.

Example of Univariant Linear regressionn any person's height & weight r dependent to each other so we can call this example as Univariant Linear Regression.
Multiple Linear regression: addition of two numnbers is an example of Multiple Linear regression ehre two independent variables & features r dependent on output or Label.
